```python
# import pandas
import pandas as pd

# Read a csv from a URL as a pandas dataframe 'df'
df = pd.read_csv('https://data.cityofnewyork.us/api/views/7yc5-fec2/rows.csv?accessType=DOWNLOAD')
```

```python
# Show information of the current dataframe

df.describe()
```

```python
# dataframe.head() method shows top 5 rows of the dataset

df.head()
```

```python
# Filter a column based on a condition (for eg., select records with Total Enrollment > 2100)

df[df['Total Enrollment'] > 2100]
```

```python
# Index - an important part of pandas dataframe
# Unless otherwise stated, pandas indexes records by default in a serial

df.index
```

```python
# Let us set index to 'Year'

df.set_index('Year')
```

```python
# Now let us retrieve the first record - just as in a list - using index 0
df.iloc[0]
```

```python
#Also, we can slice first n records just like a list
df.iloc[:4]
```

```python
# We can also slice columns using loc - below syntax
df.loc[:,'Year']
```

```python
# Or more columns as well, just pass a list having column names
df[['Year','DBN','Category']]
```

```python
# Converting a column to numpy
df['Year'].to_numpy()
```

```python
# Deriving/Making a new column
df['Year Next'] = df['Year'] + ' Next'
df['Year Next']
```
