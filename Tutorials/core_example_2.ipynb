```
import Aestheta.Library.core as core
core.getTile(show=True)
```

```
for z in range(0,20):
  core.getTile([-63.5752,44.6488,z], source = 'google_sat',show = True)
```

```
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10,5)

for z in range(0,20):
  img_RGB = core.getTile([-63.5752,44.6488,z], source = 'google_sat')
  img_class = core.classifyImage(img_RGB)

  fig, (ax1, ax2, ) = plt.subplots(1, 2)
  ax1.imshow(img_RGB);ax1.axis('off');ax1.set_title('img_RGB')
  ax2.imshow(img_class);ax2.axis('off');ax2.set_title('img_class')
  plt.show()
```
