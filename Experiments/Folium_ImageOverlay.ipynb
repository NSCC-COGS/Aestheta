```
# and set it up to be used
%cd /content/Aestheta/
!git pull
import Aestheta.Library.core as core
```

```
# now lets pull down an image using getTile()

# image=core.getTile(source='google_map')
image=core.getTile(source='google_sat')
```

```
# lets look at the type and shape of our image
print(image.dtype,image.shape)
```

```
# as it turns out, our arrays generated by getTile() are not contiguous
# this hasn't affected us so far - but seems to be a requirement for folium
# https://stackoverflow.com/questions/29947639/cheapest-way-to-get-a-numpy-array-into-c-contiguous-order/29948246

import numpy as np
image=np.ascontiguousarray(image)
```

```
# https://github.com/python-visualization/folium/blob/master/examples/ImageOverlay.ipynb
# https://github.com/python-visualization/folium/issues/709

import folium
# we could call folium to get an interactive map as such:
# we can set the start coordinates, zoom level, and tile source (and many other arguments, check out the folium docs!)
# folium.Map([45, -63], zoom_start=7, tiles='stamentoner')

# but instead, lets create a blank map object and add our array as an image overlay, then display it
# lets use bogus positioning/bounds for now
m = folium.Map([10, 50], zoom_start=5)
folium.raster_layers.ImageOverlay(image=image,bounds=[[0, 40], [20, 60]],).add_to(m)
m

#so we can see our little image showing up on an interactive map! Albeit in the wrong location.
# hmm something to work on! We should add image coordinates to getTile() :D
```
