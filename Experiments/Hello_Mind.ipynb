```
# some related info here
# https://datascience.stackexchange.com/questions/19365/predict-sinus-with-keras-feed-forward-neural-network
# https://stats.stackexchange.com/questions/307210/how-do-i-make-my-neural-network-better-at-predicting-sine-waves
# https://fairyonice.github.io/Understand-Keras's-RNN-behind-the-scenes-with-a-sin-wave-example.html
# https://machinelearningmastery.com/neural-networks-are-function-approximators/
# https://www.tensorflow.org/tutorials/structured_data/time_series

import numpy as np
from tensorflow import keras
import matplotlib.pyplot as plt
import math

# x = np.arange(0, math.pi*2, .1)
x = np.arange(0, math.pi*20, .1)
y = (np.sin(x)+1)/2

model = keras.models.Sequential([
                    keras.Input(1),
                    keras.layers.Dense(5, activation = 'sigmoid'),
                    keras.layers.Dropout(.1),
                    keras.layers.Dense(5, activation = 'relu'),
                    keras.layers.Dropout(.1),
                    keras.layers.Dense(1)
                    ])

model.compile(loss='mean_absolute_error', optimizer='adam', metrics=['mean_squared_error'])

print('fitting model...')
run = model.fit(x, y, epochs=500, batch_size=8, verbose=0, validation_split=0.2)

print('fitting complete')
```

```
from matplotlib import pyplot as plt
plt.plot(run.history['loss'], 'pink',label = 'loss')
plt.plot(run.history['val_loss'],'green', label='val loss')
plt.xlabel('epoch')
plt.legend()
plt.show()
```

```
p = model.predict(x)
plt.plot(x, y, 'b', label = 'y')
plt.plot(x, p, 'r--', label = 'p')
plt.ylabel('Y / Predicted Value')
plt.xlabel('X Value')
plt.legend()
plt.show()
```
